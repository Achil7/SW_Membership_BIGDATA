```python
# 1. flights.csv 파일 dataframe으로 읽기
import pandas as pd
df = pd.read_csv('../data/flights.csv')
df
```

```python
# 2. 결측치 파악하기. 각 열별로 결측치가 몇개 있는지 파악
df.isnull().sum()
```

```python
# 3. 결측치 대체(열 값의 평균값)
df['DEP_DELAY'].fillna(df['DEP_DELAY'].mean(), inplace= True)
df['AIR_TIME'].fillna(df['AIR_TIME'].mean(), inplace= True)
df['ARR_DELAY'].fillna(df['ARR_DELAY'].mean(), inplace= True)
df.isnull().sum()
```

```python
# 4. 항공사별(airline)로 dep_delay 값들 평균 구하기
df4 = df.groupby('AIRLINE').mean('DEP_DELAY')['DEP_DELAY']
df4 = pd.DataFrame(df4)
df4
```

```python
# 5. pivot table 이용, airline, org_air 별로 sched_dep 값들을 보는 table 만들기
# df[['AIRLINE','ORG_AIR','SCHED_DEP']]
df_sched = df.pivot_table(index = 'AIRLINE', columns= 'ORG_AIR', values= 'SCHED_DEP')
df_sched.fillna(value=0, inplace=True)
df_sched
```

```python
# 6. 4,5번 결과를 flights_new.xlsx 파일의 각 sheet로 저장
with pd.ExcelWriter('./데이터 전처리 실습 1.xlsx') as writer:
    df4.to_excel(writer, sheet_name='sheet1')
    df_sched.to_excel(writer, sheet_name='sheet2')
```
