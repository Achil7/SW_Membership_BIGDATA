# 1) 데이터 그리기
 - 데이터에 대한 기본 그래프는 라인플롯(디폴트)

```python
# 데이터 플롯
plt.plot([1,4,9,16]);
```

```python
# 쌍데이터 플롯
plt.plot([10,20,30,40],[1,4,9,16]);
```

# 그래프 꾸미기
 - 그래프 요소들


```python
plt.plot([10,20,30,40],[1,4,9,16], 'rs:');
```

```python
plt.plot([10,20,30,40],[1,4,9,16], c = 'b', lw = 5, ls = ':',
         marker='o', ms=15, mec='magenta', mew = 5, mfc = 'k');
```

```python
plt.plot([10,20,30,40],[1,4,9,16], c = 'b', lw = 5, ls = ':',
         marker='o', ms=15, mec='magenta', mew = 5, mfc = 'k');
plt.xlim(0,50)
plt.ylim(-10,30);
```

```python
# 그래프 크기 조정
plt.figure(figsize=(5,5))
plt.plot([10,20,30,40],[1,4,9,16], c = 'b', lw = 5, ls = ':',
         marker='o', ms=15, mec='magenta', mew = 5, mfc = 'k');
```

```python
# 틱 설정
import numpy as np
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.plot(x,c)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks();
```

```python
# grid 설정
# Latex 참고
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.plot(x,c)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$+\pi/$'])
plt.yticks([-1,0,1],['low','zero','high'])
plt.grid(True);
```

```python
# 중첩그래프
# 다중 라인
t = np.arange(0.5,5.,0.2)
plt.plot(t,t,'r--',t,0.5*t**2, 'bs:', t, 0.2 * t**3, 'g^-');
```

```python
# hold를 이용한 다중 라인
plt.plot([1,4,9,16], c = 'b', lw = 5, ls = '--',
         marker='o', ms=15, mec='g', mew = 5, mfc = 'r');

plt.plot([9,16,4,1], c = 'k', lw = 3, ls = ':',
         marker='s', ms=15, mec='m', mew = 5, mfc = 'c');
```

```python
x = np.linspace(-np.pi, np.pi, 256)
c,s = np.cos(x), np.sin(x)
plt.plot(x,c, ls = '--',label = 'cosine')
plt.plot(x,s, ls = ':', label = 'sine')
plt.legend(loc = 0);
```

```python
x = np.linspace(-np.pi, np.pi, 256)
c,s = np.cos(x), np.sin(x)
plt.plot(x,c, ls = '--',label = 'cosine')
plt.plot(x,s, ls = ':', label = 'sine')
plt.legend(loc = 'best');
```

```python
from matplotlib import font_manager, rc
font_fname = 'C:/windows/fonts/HANBatang.ttf'
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family=font_name)

x = np.linspace(-np.pi, np.pi, 256)
c,s = np.cos(x), np.sin(x)
plt.plot(x,c,label = 'cosine')
plt.xlabel('time')
plt.ylabel('ampitude')
plt.title('코사인 플룻')
plt.legend(loc = 0);
```

```python
# - 전체 변경 : plt.rcParams['font.family'] = 'Malgun Gothic'
# - 부분 변경 : plt.rc('font', family = 'Malgun Gothic')
```

```python
x = np.linspace(-np.pi, np.pi, 256)
c,s = np.cos(x), np.sin(x)
plt.rc('font', family = 'Malgun Gothic')
plt.plot(x,c,label = 'cosine')
plt.xlabel('time')
plt.ylabel('ampitude')
plt.title('코사인 플룻')
plt.legend(loc = 0);
```

# Pandas의 그래프


```python
import pandas as pd

df1 = pd.DataFrame(np.random.rand(5))
print(df1.head())

df2 = pd.DataFrame(np.random.rand(5))
print(df2.head())
```

```python
df1.plot() # default : line plot
df2.plot()
plt.title('랜덤 넘버 df1');
```

```python
df = pd.concat([df1,df2], axis = 1)
df.columns = ['df1','df2']
print(df)
df.plot();
```

```python
df.plot(kind = 'barh');
```

```python
df.plot(kind = 'area');
```

```python
df.plot(kind = 'scatter', x = 'df1', y = 'df2');
```

```python
fig, ax = plt.subplots()

fruits = ['apple','blueberry','cherry','orange']
counts = [40,100,30,55]
bar_labels = ['red','blue','_red','orange']
bar_colors = ['tab:red','tab:blue','tab:red','tab:orange']

ax.bar(fruits, counts, label = bar_labels, color = bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title = 'Fruit color');
```

```python
data = {'apple' : 10, 'orange' : 15, 'lemon' : 5, 'lime' :20}
names = list(data.keys())
values = list(data.values())

fig,axs = plt.subplots(1,3,figsize = (9,3), sharey = True)
axs[0].bar(names,values)
axs[1].scatter(names,values)
axs[2].plot(names,values)
fig.suptitle('Categorical Plotting');
```

```python
ts = pd.Series(np.random.randn(1000), index = pd.date_range('1/1/2000', periods = 1000))
ts = ts.cumsum()
ts.plot();
```

```python
df = pd.DataFrame(np.random.randn(1000,4),index = ts.index, columns = list('ABCD'))
df = df.cumsum()
plt.figure()
df.plot();
```
