```python
# Seaborn의 데이터셋 불러오기
```

```python
tips = sns.load_dataset('tips')
tips.head(1)
```

```python
sns.barplot(data=tips, x = 'day', y = 'total_bill');
```

```python
sns.barplot(data=tips, x= 'day', y='total_bill', hue = 'sex');
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1,2, figsize = (10,5))
sns.barplot(data=tips, x = 'day', y = 'total_bill', ax = axes[0])
sns.barplot(data=tips, x = 'day', y = 'total_bill', hue = 'sex',ax = axes[1])
```

```python
# 히스토그램
fig, axes = plt.subplots(1,2,figsize = (10,5))
sns.histplot(data=tips, x = 'total_bill', ax=axes[0])
sns.histplot(data=tips, x = 'total_bill', bins = 20, color = 'red', kde = True, ax=axes[1]);
```

```python
# 박스 플롯
fig, axes = plt.subplots(1,2,figsize = (10,5))
sns.boxplot(x=tips.smoker, y = tips.total_bill, ax = axes[0])
sns.boxplot(data = tips, x='smoker', y = 'total_bill', hue='sex', ax = axes[1]);
```

```python
# 바이올린
fig, axes = plt.subplots(1,2,figsize = (10,5))
sns.violinplot(x=tips.day, y = tips.total_bill, ax = axes[0])
sns.violinplot(data = tips, x = 'day',  y = 'total_bill', hue = 'sex', ax = axes[1]);
```

```python
print(tips.corr(method = 'pearson', numeric_only = True))

sns.heatmap(tips.corr(method = 'pearson', numeric_only=True), annot = True);
```

```python
# 다중 플롯 그리기
sns.FacetGrid(data=tips, row='sex', col = 'smoker').map(sns.histplot, 'total_bill');
```

```python
sns.jointplot(x=tips.total_bill, y= tips.tip, kind = 'scatter');
```

```python
sns.jointplot(x=tips.total_bill, y= tips.tip, kind = 'hex');
```

```python
sns.pairplot(tips);
```

```python
sns.pairplot(tips, hue='day', markers = ['o','s','d','p']);
```

```python
sns.relplot(data=tips, x ='total_bill', y='tip', hue='sex');
```

```python
sns.displot(data = tips, x = 'total_bill', col = 'time', kde = True);
```

```python
sns.displot(data = tips, kind = 'ecdf', x = 'total_bill', col = 'time', hue = 'smoker', rug = True);
```

```python
sns.lmplot(data = tips, x = 'total_bill', y = 'tip', col = 'time', hue= 'smoker');
```

```python
g = sns.PairGrid(tips, diag_sharey=False)
g.map_upper(sns.scatterplot, s=15)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw=2);
```
