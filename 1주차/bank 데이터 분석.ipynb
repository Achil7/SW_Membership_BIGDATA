```python
# 결측치 제거
print(bank_df.shape)
bank_df = bank_df.dropna(subset = ['job','education'])
print(bank_df.shape)
```

```python
# 범주형 변수 >> one-hot encoding
bank_df_job = pd.get_dummies(bank_df['job'])
bank_df_job.head()
```

```python
# 데이터 결합의 유형
A = pd.DataFrame({'ID' : [1,2,3], 'name' : ['홍길동','임꺽정','김홍익']})
B = pd.DataFrame({'ID' : [2,3,4], 'address' : ['서울','강원도','경기도']})
A,B
```

```python
# 내부 결합(Innter Join): 두 개의 테이블 키가 일치하는 데이터만 추출
inner_join = pd.merge(A,B, on = 'ID', how='inner')
inner_join
```

```python
# 좌 결합(Left Join): 왼쪽 테이블 키와 일치하는 데이터 추출
left_join = pd.merge(A,B, on = 'ID', how='left')
left_join
```

```python
# 우 결합(Right Join): 오른쪽 테이블 키와 일치하는 데이터 추출
right_join = pd.merge(A,B, on = 'ID', how='right')
right_join
```

```python
# 불균형 데이터의 처리 - 균형화
# 언더샘플링 - Random 방식
import numpy as np
from imblearn.under_sampling import RandomUnderSampler

X = np.array(bank_df_new.drop('y', axis = 1))
Y = np.array(bank_df_new[['y']])
print(np.sum(Y==1), np.sum(Y==0))

sampler = RandomUnderSampler(random_state = 1111)

X,Y = sampler.fit_resample(X,Y)
print(np.sum(Y==1),np.sum(Y==0))
```

```python
# 오버샘플링
from imblearn.over_sampling import RandomOverSampler

X = np.array(bank_df_new.drop('y', axis = 1))
Y = np.array(bank_df_new[['y']])
print(np.sum(Y==1), np.sum(Y==0))

sampler = RandomOverSampler(random_state = 1111)

X,Y = sampler.fit_resample(X,Y)
print(np.sum(Y==1),np.sum(Y==0))
```

```python
# decision-tree

from sklearn.model_selection import KFold
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score

kf = KFold(n_splits=10, shuffle=True, random_state=0)
accuracy = []
recall = []
precision = []

for train_id, test_id in kf.split(X):
    print(train_id,test_id, '\n')
    x = X[train_id]
    y = Y[train_id]
    clf = tree.DecisionTreeClassifier()
    clf.fit(x,y)
    pred_y = clf.predict(X[test_id])
    temp_accuracy = accuracy_score(Y[test_id], pred_y)
    accuracy.append(temp_accuracy)
    temp_recall = recall_score(Y[test_id], pred_y)
    recall.append(temp_recall)
    temp_precision = precision_score(Y[test_id],pred_y)
    precision.append(temp_precision)
```

```python
accuracy = np.array(accuracy)
print('Accuracy = ',accuracy.mean(), accuracy.std())

recall = np.array(recall)
print('Recall = ',recall.mean(), recall.std())

precision = np.array(precision)
print('Precision = ',precision.mean(), precision.std())
```

```python
# 하이퍼파라미터
print(clf.get_params())
```

```python
# 튜닝
# Grid Search를 이용한 파라미터 최적화
from sklearn.model_selection import GridSearchCV

params = {
    "criterion" : ["entropy"],
    "max_depth" : [40,50,60],
    "min_samples_leaf" : [1,2]}

clf_gs = GridSearchCV(tree.DecisionTreeClassifier(), params, cv = KFold(n_splits=10, shuffle=True),
                      scoring='accuracy')

clf_gs.fit(X,Y)
print(clf_gs.best_score_)
print(clf_gs.best_params_)
```
