### apply lambda 식 활용

```python
# def
def get_square(a):
    return a**2

print('3의 제곱 :', get_square(3))
```

```python
# lambda
lambda_square = lambda x: x**2
print('3의 제곱 :', lambda_square(3))
```

```python
# map
a = [1,2,3]

res = list(map(lambda x : x**2, a))
res
```

## Titanic

```python
df = pd.read_csv('./data/titanic/train.csv')
df
```

```python
# q1.
# Name의 숫자를 세어서 Na_len이라는 새로운 컬럼을 생성하여 만들라
```

```python
df['Na_len'] = df['Name'].apply(lambda x: len(x))
df
```

```python
# 컬럼 없애기 : drop
df.drop(['Na_len','Cabin'], axis = 1,inplace = True)
```

```python
# 'Age_cat' ~ 50세 이상 or 이하
df['Age_cat'] = df['Age'].apply(lambda x : 'Young_adult'  if x <= 50 else 'Elderly')
df
```

```python
['baby','child','teen','student','young_adult','adult','elderly']
```

```python
# age 5세 이하 -> 'baby'
# age 13세 이하 -> 'child'
# age 19세 이하 -> 'teen'
# age 26세 이하 -> 'student'
# age 35세 이하 -> 'young_adult'
# age 60세 이하 -> 'adult'
# 나머지 -> 'elderly'

def get_age(x):
    text = ''
    if x<= 5: text = 'baby'
    elif x<= 13 :text = 'child'
    elif x<= 19 :text = 'teen'
    elif x<= 26 :text = 'student'
    elif x<= 35 :text = 'young_adult'
    elif x<= 60 :text = 'adult'
    else: text = 'elderly'
    
    return text
        
df['Age_cat1'] = df['Age'].apply(lambda x : get_age(x))
df
```
