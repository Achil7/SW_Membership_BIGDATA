## 인덱스 객체

```python
data = {'Name' : ['Daniel','Alice','Jay','Bob'],
       'Year' : [2011,2016,2015,2015],
       'Sex' : ['Male','Female','Male','Male']}

df = pd.DataFrame(data, index = ['one','two','three','four'])
```

```python
df['Name'] # 1차원 -> Series
```

```python
df[['Name','Year']] # 2차원 -> DF
```

```python
# loc : label based indexing : 명칭 기반 인덱싱
print(df.loc['one','Name'])
df.loc['one':'two','Year':'Sex']
```

```python
# iloc : index label based indexing : 위치 기반 인덱싱
print(df.iloc[0,0])
print(df.loc[df['Year']>2014])
```
