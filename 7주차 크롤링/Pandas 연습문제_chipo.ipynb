```python
import pandas as pd
```

```python
# Q1. 다음의 chipotle데이터를 로딩해주세요. 데이터는 \t로 구분되어 있으니
# 적절한 전처리가 필요합니다.

df = pd.read_csv('../data/chipotle.tsv', sep = '\t')
df
```

```python
# Q2. 데이터의 하위 5개의 행을 출력해보세요 (tail을 활용)

df.tail(5)
```

```python
# Q3. item_name의 value값이 Chicken Salad Bowl인 df를 추출하고 index 값을 처음부터 초기화해서 적용해주세요. ^^

df[df['item_name'] == 'Chicken Salad Bowl'].reset_index(drop = True)
```

```python
# Q4. price_float의 값이 9이하이고 item_name값이 Chicken Salad Bowl인 df를 추출해주세요 ^^

df['price_float'] = df['item_price'].str[1:]  # --> $ 를 지우고 price_float 컬럼 생성

df = df.astype({'price_float':'float64'}) # --> price_float 의 dtype 변경

df[(df['price_float'] <= 9) & (df['item_name'] == 'Chicken Salad Bowl')]
```

```python
# Q5. price_float의 값을 오름차순 정리한 후 index를 새로 정의해주세요 ^^
df.sort_values(by = 'price_float').reset_index(drop = True)
```

```python
# Q6. item_name값이 Izze인 데이터를 coke로 변경해주세요. ^^

df.loc[df['item_name'] == 'Izze', 'item_name'] = 'coke'
df[df['item_name'] == 'coke']
```

```python
# Q7. df에서의 전체 결측값이 얼마인지 나타내주세요.

df.isnull().sum()#.sum()
```

```python
# Q8. df에서 choice_description 값에 Black이 들어가는 경우를 인덱싱해보세요 ^^

df[df['choice_description'].str.contains('Black', na=False)]
```

```python
# Q9. df에서 choice_description 값에 Black이 들어가지 않는 경우를 인덱싱해보세요 ^^

df[~df['choice_description'].str.contains('Black', na=False)]
```

```python
# Q10. df에서 item_name 값의 철자가 13개 이상인 데이터를 호출해보세요 ^^

df[df['item_name'].apply(lambda x: len(x) >= 13)]
```
