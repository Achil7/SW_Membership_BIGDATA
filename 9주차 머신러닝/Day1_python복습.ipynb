# Python 기초 복습

- 변수 : 데이터를 들고다니는 공간 (등호 ' = ' 기호를 사용하여, 우측의 값을 좌측의 변수명으로 선언)
- 데이터 타입 :
    - 연속형(숫자) : 정수형(int), 실수형(float)
    - 범주형(문자) : 문자열(str), 범주형(object)
    - 논리형 : 참(True) / 거짓(False) (boolean)

- 데이터 자료 구조 : 한 개 이상의 데이터 자료를 들고 다니는 공간
    - List : 대괄호[ ] / 데이터의 순서 존재, 중복허용, 추가와 삭제 및 변경 가능
    - Tuple : 소괄호( ) / 데이터의 순서 존재, 중복허용, 추가와 삭제 및 변경 불가능
    - Dictionary : 중괄호 { } / 키(key) - 값(value) 으로 구성됨, 추가와 삭제 및 변경 가능
    - Set(집합) : 중광호 { } / 데이터의 순서나 중복 없음, 추가와 삭제 및 변경 가능

- 불러온 01_Data.csv 파일로 아래의 미션을 수행하고자 한다.

```python
# 1. 신규고객에 대한 특성을 분석하고자 한다.
#     - 데이터에서, Sales_Type에 '렌탈'(신규고객) 해당하는 고객추출, df_rental로 선언
#     - df_rental에서 Channel 유입 경로 별 고객의 Credit_Rank신용등급의 평균을 요약 계산하시오.

df_rental = df[df['Sales_Type'] == '렌탈']
df_rental.pivot_table(index = 'Channel', values = 'Credit_Rank', aggfunc = 'mean')
```

```python
# 2. 재계약 고객의 특성을 확인하고자 한다.
#     - 데이터에서, Sales_Type이 '멤버십'인 고객들을 추출한 뒤, Amount_Month가 높은 상위 100명의 고객 명단을 추출하여 result2.xlsx 저장

df[df['Sales_Type'] == '멤버십'].sort_values(by = 'Amount_Month', ascending = False).head(100)#.to_excel('./실습 저장파일/result2.xlsx')
```

```python
# 3. 불량 고객의 특성을 확인하고자 한다.
#     - 데이터에서, Overdue_count 연체건수가 1건 이상인 고객들을 뽑아, Age 연령이 높은 순서대로 정렬해, df_overdue이름의 변수로 선언
#     - df_overdue에서, Product_Type 제품군 별 Overdue_count 연체건수의 평균을 계산하여 result3.xlsx 저장
    
df_overdue = df[df['Overdue_count'] >= 1].sort_values(by = 'Age', ascending = False)
df_overdue.pivot_table(index = 'Product_Type', values = 'Overdue_count', aggfunc = 'mean')#.to_excel('./실습 저장파일/result3.xlsx')
```

# Mission 2.

- 데이터로 아래의 과제를 수행하고자 한다

```python
# 1. 유입경로(Channel)에서 '영업방판'으로 유입된 고객들을 분석하고자 한다.
#'영업방판'으로 유입된 고객들의 월 랜탈 비용(Amount_Month)의 평균과 중앙값의 차이를 계산하시오.

a = df.loc[df['Channel'] == '영업방판']['Amount_Month'].mean()
b = df.loc[df['Channel'] == '영업방판']['Amount_Month'].median()
print('평균 :',a)
print('중앙값 :',b)
a - b
```

```python
# 2. 유입경로(Channel)에서 '영업방판'으로 유입된 고객들 중
# 제품군(Product_Type)별 월랜탈비용(Amount_Month)의 합을 계산하여 result4.xlsx로 저장하시오.

df_channel = df.loc[df['Channel'] == '영업방판']
df_channel.pivot_table(index = 'Product_Type', values = 'Amount_Month', aggfunc = 'sum')#.to_excel('./실습 저장파일/result4.xlsx')
```

```python
# 3. 유입경로(Channel)에서 '대형마트A'로 유입된 고객들 중
# 연체건수(Overdue_count)가 높은 상위 30개의 계약 명단을 추출하여 result5.xlsx로 저장하시오.

df.loc[df['Channel'] == '대형마트A'].sort_values(by = 'Overdue_count', ascending = False).head(30)#.to_excel('./실습 저장파일/result5.xlsx')
```

```python
# 4. 월랜탈비용(Amount_Month)이 15만원 이상인 고객들 중
# 판매유형(Sales_Type)이 '렌탈'에 해당하는 고객들만 추출하여, 해당 데이터의 연속형 기술 통계량을 확인하시오.
df_150000 = df.loc[df['Amount_Month'] >= 150000]
df_150000_렌탈 = df_150000.loc[df_150000['Sales_Type'] == '렌탈']
df_150000_렌탈.describe()
```

```python
# 5. 제품군(Product_Type)중 빈도수가 높은 상위 3개 항목을 추출하여
# df3으로 선언한 뒤, 계약일자(Datetime)에 따른 월랜탈비용(Amount_Month)의 합을 계산해, result6.xlsx로 저장 하시오.

df3 = df.loc[df['Product_Type'].isin(df['Product_Type'].value_counts().head(3).index)]
df3.pivot_table(index = 'Datetime', values = 'Amount_Month', aggfunc = 'sum')#.to_excel('./실습 저장파일/result6.xlsx')
```

```python
# 6. 납부유형(Payment_Type)에서 'CMS'에 해당하는 데이터들을 추출하여
# df4로 선언하고 df4에서 성별(Gender)에 따른 연체건수(Overdue_count)의 평균과 합을 계산하시오. 

df4 = df.loc[df['Payment_Type'] == 'CMS']
df4.pivot_table(index = 'Gender', values = 'Overdue_count', aggfunc = ['mean','sum'])
```
